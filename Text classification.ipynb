```python
import pandas as pd
```

```python
df = pd.read_csv("Back_Statements_data1.csv",names=['sent',"type"])
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["tr_type"] = le.fit_transform(df["type"])
```

```python
df
```

```python
df.drop(columns ="type",inplace=True)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
data = cv.fit_transform(df["sent"]).toarray()
data
```

```python
features = cv.get_feature_names_out()
```

```python
features
```

```python
new_df = pd.DataFrame(data,columns=features)
```

```python
new_df
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(new_df,df["tr_type"])
```

```python
r = cv.transform([s]).toarray()
```
